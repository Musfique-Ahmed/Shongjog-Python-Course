Handleling Missing Values

## Dropping Missing Values

*   This approach removes any rows or columns containing missing data.
*   Useful when missing data is minimal and won't significantly impact the dataset.



Axis = 0 -> Row
Axis = 1 -> Column

```python
# Droping Rows

df_dropped_rows = df.dropna().reset_index()
df_dropped_rows
```

```python
df_dropped_col = df.dropna(axis=1).reset_index()
df_dropped_col
```
