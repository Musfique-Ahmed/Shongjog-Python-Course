## Pie Chart

## Customization

```python
x = [1,2,3,4,5]
y = [10, 20, 25, 15, 60]
z = [69 , 50 , 33 , 27 , 14]

plt.plot(x, y, label = "Line 1", color = "Purple", linestyle = "--", marker = "o")
plt.plot(x, z, label = "Line 2", color = "red", linestyle = "--", marker = "o")
plt.xlabel("x-axis")
plt.ylabel("Y-axis")
plt.title("Customized Line Plot")
plt.legend()
plt.grid(True)
plt.annotate("Highest Values", xy = (5,50), xytext=(3, 40), arrowprops=dict(facecolor = 'black'))
plt.show()
```

## SubPlots

```python
fig, axs = plt.subplots(2,2, figsize = (8,8))

x = [1,2,3,4,5]
y = [10, 20, 25, 15, 60]

axs[0,0].plot(x,y, color = 'blue')
axs[0,0].set_title("Line Plot")

axs[0,1].bar(x,y, color = 'red')
axs[0,1].set_title("Bar Plot")

axs[1,0].scatter(x,y, color = 'green')
axs[1,0].set_title("Scatter Plot")

axs[1,1].hist(y,bins=5, color = 'blue')
axs[1,1].set_title("Histogram")

plt.tight_layout()
plt.show()
```

## Real-World Data Visualization

```python
import pandas as pd

df = pd.DataFrame({
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Sales': [200, 250, 300, 350, 400]
})

plt.plot(df['Year'], df['Sales'], marker='o', linestyle='-', 
         color='blue')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Over Time')
plt.show()
```
