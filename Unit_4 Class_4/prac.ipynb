## Line Plot

```python
x = [1,2,3,4,5]
y = [10, 20, 25, 15, 60]

plt.plot(x, y)
plt.xlabel("x-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.show()
```

```python
x = ["Ahnaf" , "Raisa" , "Wasi" , "Manha" , "Tafsir", "Abdullah"]
y = [14 , 27 , 33 , 49 , 50 , 69]
z = [69 , 50 , 49 , 33 , 27 , 14]
plt.plot(x , y)
plt.plot(x , z)
plt.xlabel("student improvement report")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.show()
```

```python
df_copy = df.copy()
sales_count = df_copy.groupby("Product")["Sales"].sum()
sales_count
```

```python
plt.plot(sales_count)
plt.xlabel("Product Name")
plt.ylabel("Quantity")
plt.title("Sales Report")
plt.show()
```

## Scatter Plot

```python
x= np.random.rand(50)
y= np.random.rand(50)

plt.scatter(x,y, color="red", marker="p")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

```python
height = df["Height"]
weight= df["Weight"]
plt.scatter(height,weight, color="red",marker="o")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Height vs Weight")
plt.show()
```

```python
x = df["Height"]
y = df["Weight"]

plt.scatter(x, y, color = "blue", marker= "x")
plt.scatter(df["Market_Share"], y, color="red", marker="v")
plt.xlabel("Height of the Students")
plt.ylabel("Weight of the Students")
plt.title("Height vs Weight")
plt.show()
```

## Bar Chart

```python
cat = ["A", "B", "C", "D"]
val = [24, 35, 13, 46]

plt.bar(cat, val, color = "green")
plt.xlabel("Category")
plt.ylabel("Values")
plt.title("Bar Chart")
plt.show()
```

```python
x = df["Product"]
y = df["Sales"]

plt.bar(x,y,color="red")
plt.xlabel("Product Name")
plt.ylabel("Quantity")
plt.title("Sales Report")
plt.show()
```

```python
cat = ["A", "B", "C", "D", "E"]
val = [1104,420,2268,1296,1938]

plt.bar(cat, val, color="green")
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.title("Bar Chart")
plt.show()
```

```python
df_copy = df.copy()
sales_count = df_copy.groupby("Product")["Sales"].sum().reset_index()
sales_count
```
