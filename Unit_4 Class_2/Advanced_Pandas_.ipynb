# Group By Mechanism
The groupby function in pandas is a very powerful and versatile tool for splitting data into groups, applying a function to each group, and combining the results. This process can be broken down into three steps: splitting, applying, and combining.

The groupby operation involves:

-Splitting the data into groups based on some criteria.

-Applying a function to each group independently.

-Combining the results into a data structure.
Picture1.png


```python
import pandas as pd

data = {
    'Name': ['Tamanna', 'Faysal', 'Tarek', 'Raisa', 'Imtiaz', 'Farah'],
    'Age': [24, 27, 22, 32, 24, 22],
    'City': ['Dhaka', 'Sylhet', 'Dhaka', 'Khulna', 'Rajshahi', 'Sylhet'],
    'Score': [85, 95, 78, 85, 88, 84]
}

df = pd.DataFrame(data)
grouped = df.groupby('City')
print(grouped)

mean_scores = grouped['Score'].mean()
print(mean_scores)
```

# Iterating Over Groups

```python
for name, group in grouped:
    print(f"City: {name}")
    print(group)
    print()
```

Multiple column grouping

```python
data = {
    'Name': ['Tamanna', 'Faysal', 'Tarek', 'Raisa',
             'Imtiaz', 'Farah','Tamanna'],
    'Age': [24, 27, 22, 32, 24, 22,21],
    'City': ['Dhaka', 'Sylhet', 'Dhaka', 'Khulna', 'Rajshahi',
             'Sylhet','Dhaka'],
    'Score': [85, 95, 78, 85, 88, 84,90]
}

df = pd.DataFrame(data)
grouped=df['Age'].groupby([df['Name'],df['City']]).mean()
print(grouped)
```

# Data Aggregation
Aggregations refer to any data transformation that produces scalar values from arrays. The preceding examples have used several of them, including mean, count, min, and sum.
image.png

What if we want to aggregate based on our own function?

```python
# Define a custom function to calculate the range of scores
def score_range(group):
    return group.max() - group.min()

# Apply the custom function to each group
score_ranges = df.groupby('City')['Score'].apply(score_range)
print(score_ranges)
```

```python
# Group by 'City' and apply multiple aggregation functions
agg_results = df.groupby('City')['Score'].agg(['mean', 'max', 'min'])
print(agg_results)
```

```python
# Group by 'City' and apply multiple aggregations
agg_result = df.groupby('City')['Score'].agg(['mean', 'max',
                                              score_range])
print(agg_result)
```

Quiz: What happened here?

```python
# Group by 'City' and apply multiple aggregation functions to 'Age' and 'Score'
functions=['count','max','min']
result = df.groupby('City')[['Age', 'Score']].agg(functions)
print(result)
```

Suppose you want to apply potentially different functions to one or more of your columns

```python
result = df.groupby('City').agg({'City':'count','Age':'mean'})
print(result)
```

# Apply

In pandas, the apply function is used to apply a function along an axis of a DataFrame or specific groups within a DataFrame.

```python
def top(group):
    return group.nlargest(2, 'Score')

result = df.groupby('Age').apply(top)

print(result)
```

Apply - To Fill Missing Values

```python
import numpy as np
import pandas as pd

data = {
    'Name': ['Tamanna',  'Tarek', 'Raisa', 'Imtiaz', 'Farah', 'Tamanna'],
    'Age': [24,  22, 32, 24, 22, np.nan],
    'City': ['Dhaka', 'Dhaka', 'Khulna', 'Rajshahi', 'Sylhet', 'Dhaka'],
    'Score': [85,  78, 85, 88, 84, 90]
}


df = pd.DataFrame(data)

# Define a function to fill missing values with the mean of the group
def fill_missing_with_mean(group):
    mean_age = group['Age'].mean()  # Calculate the mean age for the group
    group['Age'] = group['Age'].fillna(mean_age)  # Fill missing Age values with the mean
    return group

df = df.groupby('City').apply(fill_missing_with_mean).reset_index(drop=True)
print(df)
```

## Categorical Values

	Categorical/dictionary-encoded representation: using integers to identify unique values

	The array of distinct values is called categories

	The integer values that represent the categories are called codes

	This yields significant performance improvements and is a better representation of data for models

Color (Original) | Code (Encoded)
----------------|----------------
Red             | 0
Blue            | 1
Red             | 0
Green           | 2
Blue            | 1




```python
import pandas as pd

# Original DataFrame
df = pd.DataFrame({'Color': ['Red', 'Blue', 'Red', 'Green', 'Blue']})

# Convert the 'Color' column to a categorical type
df['Color'] = df['Color'].astype('category')
#color=df['Color'].astype('category')
#print(color.values)
# Get the categories (distinct values)
categories = df['Color'].cat.categories
print("Categories:", categories)

# Get the codes (integer representation)
codes = df['Color'].cat.codes
print("Codes:\n", codes)

# Add the codes to the DataFrame
df['Color_Code'] = codes
print("\nEncoded DataFrame:\n", df)
```

If you know the codes from elsewhere, you can use them to encode values

```python
categories=['Red','Green','Blue']
codes=[0,1,2,0,0,1,2]

cat_2=pd.Categorical.from_codes(codes,categories)
print(cat_2)
```
